```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga desde un archivo .xlsx sin indice
df = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx')
df
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
#FILTROS BÁSICOS PARA VARIABLES CUALITATIVAS*
```

```python
#Filtro por objeto
filtro1=df[df['NOMBRE_VENDEDOR']=='LETICIA RAMIREZ HERNANDEZ']
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Actividad_1.1.csv")
```

```python
#filtro por filas
#filtro 2 =df.iloc[3:10, :]  #Filas consecutivas
filtro2 =df.iloc[ [8,24,100], : ]
filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("Actividad_1.2.csv")
```

```python
# Filtro por columnas
filtro3 = df.iloc[ : , [2,3,7]] # Columnas no consecutivas
filtro3
```

```python
#Convertir archivo filtrado a CSV
filtro3.to_csv("Actividad_1.3.csv")
```

```python
df1 = pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col =2)
df1
```

```python
filtro4 = df1.loc[["ALFREDO CANELA", "LETICIA RAMIREZ HERNANDEZ"], ["COSTO_TOTAL_CALCULADO"]]
filtro4
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv("Actividad_1.4.csv")
```

```python
#Filtro por cola
filtro5 = df.tail(10)
filtro5
```

```python
#Convertir archivo filtrado a CSV
filtro5.to_csv("Actividad_1.5.csv")
```

```python
#Filtro para Cabecera
filtro6=df.head(10)
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("Actividad_1.6.csv")
```

```python
#Filtro por comparación
filtro7=df[df['COSTO_TOTAL_CALCULADO'] > 900]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro7.to_csv("Actividad_1.7.csv")
```

```python
#Filtro Y
filtro8=df[(df["COSTO_TOTAL_CALCULADO"]<400) & (df["FECHA_DOC"]=="2022-01-05")]
filtro8
```

```python
#Convertir archivo filtrado a CSV
filtro8.to_csv("Actividad_1.8.csv")
```

```python
# filtro o
filtro9=df[(df["COSTO_TOTAL_CALCULADO"] < 600)| (df["FECHA_DOC"]=="2023-01-05")]
filtro9
```

```python
#Convertir archivo filtrado a CSV
filtro9.to_csv("Actividad_1.9.csv")
```

```python
#filtro not
filtro10=df[~(df["COSTO_TOTAL_CALCULADO"] > 700) & ~(df["FECHA_DOC"]=="2023-01-05")]
filtro10
```

```python
#Convertir archivo filtrado a CSV
filtro10.to_csv("Actividad_1.10.csv")
```
